```python
import pandas as pd
url ="/content/iris.csv"
data = pd.read_csv(url)
data.head()
```

```python
import matplotlib.pyplot as plt 
data.hist()
plt.show
```

```python
correlation = data.corr()
print (correlation )
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
```

```python
sns.pairplot(data)
```

```python
from sklearn import linear_model #hoi quy tuyen tinh
model =linear_model.LinearRegression()
model.fit(X=pd.DataFrame (data['petal.length']),
          y=data['sepal.length'])
print ("He so intercept la :", model.intercept_) 
print ("He so coeffient la:", model.coef_)
model.score(X=pd.DataFrame (data['petal.length']),
          y=data['sepal.length'])# tinh do chinh xac
```

```python
y_pred=model.predict (X=pd.DataFrame (data['petal.length']))
print(y_pred)
```

```python
data.plot (kind= "scatter",x="petal.length",
           y= "sepal.length", figsize=(9,9),
           color = "black", xlim=(0,7))
plt.plot(data["petal.length"],y_pred,color ="purple")
```

```python
model =linear_model.LinearRegression()
X = pd.DataFrame([data["petal.length"],data["petal.length"]**2]).T
y= data["sepal.length"]
model.fit (X,y)
print ('Intercept cua mo hinh :', model.intercept_)
print ("Coeffient cua mo hinh :", model.coef_)
print ('Do chinh xac cua mo hinh :', model.score(X,y))
```

```python
import numpy as np 
range = np.arange (0.5,10,0.1)
X_range = pd.DataFrame ([range , range **2]).T
y_pred=model.predict (X=X_range)
data.plot (kind="scatter",x="petal.length",
           y= "sepal.length",figsize =(9,9),color='black',
           xlim=(0,10))
plt.plot(range , y_pred,color="blue")
```

```python
model =linear_model.LinearRegression()
X = pd.DataFrame([data["petal.length"],data["petal.length"]**2,data["petal.length"]**3]).T
y= data["sepal.length"]
model.fit (X,y)
print ('Intercept cua mo hinh :', model.intercept_)
print ("Coeffient cua mo hinh :", model.coef_)
print ('Do chinh xac cua mo hinh :', model.score(X,y))
```

```python
import numpy as np 
range = np.arange (0.5,10,0.1)
X_range = pd.DataFrame ([range , range **2,range**3]).T
y_pred=model.predict (X=X_range)
data.plot (kind="scatter",x="petal.length",
           y= "sepal.length",figsize =(9,9),color='black',
           xlim=(0,10))
plt.plot(range , y_pred,color="orange")
```
