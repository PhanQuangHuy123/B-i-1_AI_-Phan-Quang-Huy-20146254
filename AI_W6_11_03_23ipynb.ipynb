```python
import numpy as np 
import skfuzzy as fuzz
from skfuzzy import control as ctrl
setpoint = ctrl.Antecedent(np.arange(18,27,0.1),"setpoint")
vhl_sensor = ctrl.Antecedent(np.arange(0,40,0.1),"vhl_sensor")
ac = ctrl.Consequent(np.arange(0,60,1),"ac")

setpoint['verylow']= fuzz.trimf(setpoint.universe,[18,19,20])
setpoint['low']= fuzz.trimf(setpoint.universe,[19,20,21])
setpoint['normal']=fuzz.trimf(setpoint.universe,[20,21,23])
setpoint['high']= fuzz.trimf(setpoint.universe,[22,24,25])
setpoint['veryhigh']= fuzz.trimf(setpoint.universe,[24,25,27])

setpoint.view()


vhl_sensor['verylow']= fuzz.trimf(vhl_sensor.universe,[0,5,10])
vhl_sensor['low']= fuzz.trimf(vhl_sensor.universe,[5,10,15])
vhl_sensor['normal']=fuzz.trimf(vhl_sensor.universe,[12,20,25])
vhl_sensor['high']= fuzz.trimf(vhl_sensor.universe,[20,30,35])
vhl_sensor['veryhigh']= fuzz.trimf(vhl_sensor.universe,[25,35,40])

vhl_sensor.view()


ac['verylow']= fuzz.trimf(ac.universe,[0,5,10])
ac['low']= fuzz.trimf(ac.universe,[5,10,15])
ac['normal']=fuzz.trimf(ac.universe,[15,25,40])
ac['high']= fuzz.trimf(ac.universe,[20,40,50])
ac['veryhigh']= fuzz.trimf(ac.universe,[30,50,60])

ac.view()

rule1 = ctrl.Rule(setpoint['verylow']&vhl_sensor['verylow'],ac['verylow']) 
rule2 = ctrl.Rule(setpoint['verylow']&vhl_sensor['low'],ac['verylow'])  
rule3 = ctrl.Rule(setpoint['verylow']&vhl_sensor['normal'],ac['low'])
rule4 = ctrl.Rule(setpoint['verylow']&vhl_sensor['high'],ac['normal'])
rule5 = ctrl.Rule(setpoint['verylow']&vhl_sensor['veryhigh'],ac['normal'])

rule6 = ctrl.Rule(setpoint['low']&vhl_sensor['verylow'],ac['verylow'])
rule7 = ctrl.Rule(setpoint['low']&vhl_sensor['low'],ac['low'])
rule8 = ctrl.Rule(setpoint['low']&vhl_sensor['normal'],ac['normal'])
rule9 = ctrl.Rule(setpoint['low']&vhl_sensor['high'],ac['normal'])
rule10 = ctrl.Rule(setpoint['low']&vhl_sensor['veryhigh'],ac['high'])

rule11 = ctrl.Rule(setpoint['normal']&vhl_sensor['verylow'],ac['low'])
rule12 = ctrl.Rule(setpoint['normal']&vhl_sensor['low'],ac['normal'])
rule13 = ctrl.Rule(setpoint['normal']&vhl_sensor['high'],ac['high'])
rule14 = ctrl.Rule(setpoint['normal']&vhl_sensor['high'],ac['high'])
rule15 = ctrl.Rule(setpoint['normal']&vhl_sensor['veryhigh'],ac['high'])

rule16 = ctrl.Rule(setpoint['high']&vhl_sensor['verylow'],ac['normal'])
rule17 = ctrl.Rule(setpoint['high']&vhl_sensor['low'],ac['normal'])
rule18 = ctrl.Rule(setpoint['high']&vhl_sensor['normal'],ac['high'])
rule19 = ctrl.Rule(setpoint['high']&vhl_sensor['high'],ac['veryhigh'])
rule20 = ctrl.Rule(setpoint['high']&vhl_sensor['veryhigh'],ac['veryhigh'])

rule21 = ctrl.Rule(setpoint['veryhigh']&vhl_sensor['verylow'],ac['normal'])
rule22 = ctrl.Rule(setpoint['veryhigh']&vhl_sensor['low'],ac['high'])
rule23 = ctrl.Rule(setpoint['veryhigh']&vhl_sensor['normal'],ac['high'])
rule24 = ctrl.Rule(setpoint['veryhigh']&vhl_sensor['high'],ac['veryhigh'])
rule25 = ctrl.Rule(setpoint['veryhigh']&vhl_sensor['veryhigh'],ac['veryhigh'])


acing_ctrl= ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
acing  = ctrl.ControlSystemSimulation(acing_ctrl)
acing.input['setpoint']=25.5
acing.input['vhl_sensor']=20.25
acing .compute()
print(acing.output['ac'])
ac.view (sim= acing)
```
